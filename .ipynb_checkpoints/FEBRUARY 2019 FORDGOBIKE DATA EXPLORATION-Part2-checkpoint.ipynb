## Investigation Overview

We perform this analysis to uncover the prevalence of the bike sharing system and factors that may affect it.

## Dataset Overview

This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('my_data.csv')
```

## A plot to show how the gender is distributen across duration and datehour
!
### This slide establishes the fact that a vast majority of people do not share their bike but it however highlights a few points. This shows that thereis a fairly equal amount of genders who either share or do not share the bike rides. But in general, it can be noted that men share the bikes most.


```python
g=sns.catplot(data=df,kind="count", x="member_gender",height=8,
           hue="bike_share_for_all_trip")
g.fig.suptitle('gender sharing status')
```

## Bike sharing options for users

#### This chart however shows clearly that customers do not use the bike sharing option at all and a percentaage of the subscribers use the sharing option.

```python
p=sns.catplot(data=df,kind="count", hue="bike_share_for_all_trip", x="user_type",height=8)
p.fig.suptitle('user_type vs bike_sharing')
```

## Bike traffic time chart

#### This chart accurately shows the time of the day with the highest customer usage.  From the chart, we can observe that 7-9am and 4-6pm see the most traffic of riders

```python
plt.figure(figsize = [10, 7])
sns.countplot(x='datehour',data=df,palette='Blues').set_title('Time count for times people use the bike most')
plt.grid(b=None)
```

## User type duration

#### We pbserve here that customers ride for longer periods than subscribers

```python
plt.figure(figsize = [10,7])
sns.barplot(x='user_type',y='duration_mins',data=df,palette='Blues').set_title('B3:user type ride duration')
```

# Facet grid

### This is a multivariate chart to demonstrate how member gender relates to user type and how they are affected by member age and duration.

#### The "other" gender has the least values but with higher durations so they could be considered at outliers
#### There are more male riders than female riders
#### The duration reducs with age therefore, the older the rider is, the less time he/she is going to ride for
#### There is little or no difference between theh the distribution of customers and subscribers

```python
g.fig.set_size_inches(20, 8)
g = sns.FacetGrid(df, row="member_gender",  col="user_type")
g.map(sns.scatterplot, "member_age", "duration_mins")
g.fig.suptitle('duration vs member age')
g.fig.tight_layout(w_pad=18)
```
